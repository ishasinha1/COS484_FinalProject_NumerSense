```python
#VerbPhysics dataset
df1 = pd.read_csv('train.csv')
df2 = pd.read_csv('dev.csv')
df3 = pd.read_csv('test.csv')
frames = [df1, df2, df3]
#combine
data = pd.concat(frames)
data = data.drop(['Unnamed: 0'], axis=1)
```

```python
data = data.reset_index()
data = data.drop(['index'], axis=1)
data.head()
```

```python
data
```

```python
#go through all of the data
for idx in range(0, len(data)):
  obj1 = data['obj1'][idx]
  obj2 = data['obj2'][idx]
  #replace any ambiguos words with words of similar magnitude
  if (obj1 =="master"):
    obj1 = "father"
  if (obj2 =="master"):
    obj2 = "father"
  if (obj1 =="servant"):
    obj1 = "child"
  if (obj2 =="servant"):
    obj2 = "child"
  if (obj1 =="air"):
    obj1 = "universe"
  if (obj2 =="air"):
    obj2 = "universe"
  #for theoretical comparisons delete
  if (obj2 =="heaven" or obj1 == "heaven" or obj1 == "soul" or obj2 =="soul"):
    data = data.drop(labels=idx, axis=0)
    continue
  #the researchers manually labeled any disagreements or wrong comparisons with -42
  if (data['size-maj'][idx]==-42):
    data = data.drop(labels=idx, axis=0)
    continue
  #if it is -1 than obj1 is smaller
  if (data['size-maj'][idx]==-1):
    data['speed-agree'][idx] = obj1 + " is smaller than " +obj2
    data['speed-maj'][idx] = "smaller"
  #if it is 1 than obj1 is larger
  else:
    data['speed-agree'][idx] = obj1 + " is larger than " +obj2
    data['speed-maj'][idx] = "larger"
```

```python
#clean the df
data = data.rename(columns={"speed-agree": "text", "speed-maj": "key"})
data = data.reset_index()
data = data.drop(['index', 'obj1', 'obj2', 'size-agree', 'size-maj','weight-agree' , 'weight-maj', 'strength-agree', 'strength-maj', 'rigidness-agree', 'rigidness-maj'], axis=1)
data.head(10)
```

```python
data = data.drop(['level_0'], axis=1)
#save and manually check
filename = 'sizeOrig.csv'

data.to_csv('/content/drive/My Drive/' + filename)
```

```python
#code for masking and lower case
for idx in range(len(data)):
    key = data['key'][idx]
    data['text'][idx] = data['text'][idx].lower()
    data['text'][idx] = data['text'][idx].replace(key, '<mask>')  # Convert to lowercase.
```
