```python
training_dataset = dataset['train']
total_examples = len(training_dataset)
cities_usa = [city.lower() for city in city_names]
```

```python
num_usa = 0
for example in tqdm(training_dataset):
  text = example['text'].lower()
  if 'united states' in text:
    num_usa += 1
    continue
  for city in cities_usa:
    if city in text:
      num_usa += 1
      break
```

```python
print('total number of examples:', total_examples)
print('examples that mention the united states/some part of it:', num_usa)
print('% of examples that mention the united states/some part of it:', num_usa / total_examples)
```
